**CONCEPTOS BASICOS DE LA LIBRERIA MATPLOTLIB**

Matplotlib es una libreria de trasados de graficos de bajo nivel en python que sirve como una herramienta de visualizacion. Matplotlib es e codigo abierto y podemos utilizarlo libremente. Matplotlib esta escrito principalmente en python con algunos segmentos.


**INSTALACION DE MATPLOTLIB **


si tienes python y pip ya instalado en ru sistema, instala matplotlib usando el siguiente comando:
pip install matplotlib desde la linea de comando.


**IMPORTAR MATPLOTLIB **

Una vez instalado matplotlib, importarlo en tus aplicaciones agregando el modulo import, ejemplo:
import matplotlib
