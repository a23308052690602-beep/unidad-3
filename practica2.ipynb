plottin sin linea. para trazar solo los marcadores, puede usar notacion de cadena  de accaso directo 'o', que significa "anillos". ejemplo:dibuja dos puntos en el diagrama, uno en la posicion (1,3) y otro en la posicion (8,10)

```python
import matplotlib.pyplot as plt
import numpy as np

xpoints=np.array([1,8])
ypoints=np.array([3,10])

plt.plot(xpoints, ypoints, 'o')
plt.show()
```

diagramas de dispersion (scratter)

 se emplean para mostrar la relacion entre  dos variables numericas, detectando correlaciones o agrupamiento, ejemplo:



```python
import matplotlib.pyplot as plt

altura=[150, 160, 170, 180, 190, 200]
peso=[50, 60, 65, 80, 85, 95]

plt.scatter(altura, peso, color = 'red', marker ='x')
plt.title("relacion entre la altura y el peso")
plt.xlabel("altura (cms)")
plt.ylabel("peso (kg)")
plt.show()
```

graficos de barras (bar y barh)

sirven para comparar categorias o variables tanto en orientacion vertical como horizonta.

ejemplo de barras verticales:

```python
import matplotlib.pyplot as plt

categorias =["a", "b", "c", "d"]
valores=[4, 7, 1, 8]

plt.bar(categorias, valores, color= "orange")
plt.title("grafica de barras")
plt.xlabel("categorias")
plt.ylabel("valores")
plt.show()
```
